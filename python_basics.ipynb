```python
# prompt: i want to generate mean, median, standard deviation from salary column

# Mean of the 'salary' column
mean_salary = df['salary'].mean()
print(f"Mean Salary: {mean_salary}")

# Median of the 'salary' column
median_salary = df['salary'].median()
print(f"Median Salary: {median_salary}")

# Standard deviation of the 'salary' column
std_dev_salary = df['salary'].std()
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT Score')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores')
plt.show()
```

```python
sns.scatterplot(data=df1, y='SAT', x=df1.index,color='brown')
plt.xlabel('Index')
plt.ylabel('SAT Score')
plt.title('Scatter Plot of SAT Scores')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.show()
```

```python
sns.histplot(df1['Accept'],color='orange')
plt.xlabel('Acceptance Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Acceptance Rates')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SAT Score')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rates')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'go')
plt.xlabel('SAT Score')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rates')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'bv')
plt.xlabel('SAT Score')
plt.ylabel('Top 10% High School Grad Rate')
plt.title('Scatter Plot of SAT Scores vs. Top 10%')
plt.show()
```

```python
sns.pairplot(df1)
plt.show()
```

```python
sns.pairplot(df1.iloc[:,1:3])
plt.show()
```

```python
plt.boxplot(df1['SAT'])
plt.show()
```
